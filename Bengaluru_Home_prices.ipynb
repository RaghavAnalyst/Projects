we can see that there are range values also present in the dataset. Therefore, we need to remove them as well

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
'''
The above function states that whenever it will see a float number, it will return the same. Otherwise it shall remain the other mentioned numbers
'''
```

```python
new_dataset[~new_dataset['total_sqft'].apply(is_float)] #   ~ indicates the opposite of the condition mentioned
```

```python
new_dataset[~new_dataset['total_sqft'].apply(is_float)].head(10)
```

```python
# we can apply a function to find the average of the numbers and use the same
```

```python
def convert_sqft_to_num(x):
    tokens = x.split("-")
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
            
```

```python
convert_sqft_to_num('1195 - 1440')
```

```python

new_dataset['total_sqft'] = new_dataset['total_sqft'].apply(convert_sqft_to_num)
```

```python
new_dataset.head(20)
```

```python
new_dataset.isnull().sum()
```

```python
nd_mean = new_dataset['total_sqft'].mean()
```

```python
new_dataset['total_sqft'] = new_dataset['total_sqft'].fillna(nd_mean)
```

As this is the string type data we have to convert this into integer form using label encoder but we cannot use so many rows for the same.

For that purpose, we will check the locations with highest frequencies and then for remaining we shall write others

```python
new_dataset1['location'] = new_dataset1['location'].apply(lambda x:x.strip())# to remove additional spaces in the dataset
location_stats = new_dataset1.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

We shall mention all the rows with values less than 10 as others 

```python
len(location_stats[location_stats <= 10])
```

```python
location_stats_less_than_10 = location_stats[location_stats <= 10]
location_stats_less_than_10
```

```python
new_dataset1['location'] = new_dataset1['location'].apply(lambda x: 'Others' if x in location_stats_less_than_10 else x)
```
