# Exploratory Data Analytics

```python
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
sns.barplot(x=df['Region'],y=df['Profit'],hue=df['Segment'])
```

# Insights

1. South has highest profit in Consumer segment
2. West has highest profit in Corporate segment
3. Central has highest profit in Home Office segment
4. East has highest profit in Home Office segment

```python
sub=df.groupby('Ship Mode')['Sales','Profit'].mean()
sub
```

```python
sub.plot(x='Ship Mode',y='Sales',
         kind='pie',radius=1.5,
         autopct="%f%%",
         labels=['First Class','Same Day','Second Class','Standard Class'])
plt.legend(loc='center right')
plt.title('Sales Contribution')
```

```python
sub.plot(x='Ship Mode',y='Profit',
         kind='pie',radius=1.5,
         autopct="%f%%",
         labels=['First Class','Same Day','Second Class','Standard Class'])
plt.legend(loc='center right')
plt.title('Profit Contribution')
```

# Finding insights and patters from the dataset

```python
profit=df.groupby('State')['Profit'].sum()
profit
```

```python
print(f'The State with highest profit is {profit.idxmax()} and the profit numbers are {profit.max()}' )
print(f'The State with highest loss is {profit.idxmin()} and the loss numbers are {profit.min()}' )
```

```python
plt.figure(figsize=(15,20))
sns.set_style('darkgrid')
profit.plot(kind='barh')
plt.xlabel('Profit')
plt.ylabel('State')
```

# Insights

Oregan and Florida are suffering loss but still less in comparison to Texas

```python
sales=df.groupby('State')['Sales'].sum()
sales
```

```python
print(f'The State with the highest number of sales is {sales.idxmax()} and the numbers are {sales.max()}')
print(f'The State with the least number of sales is {sales.idxmin()} and the numbers are {sales.min()}')
```

```python
sales_versus_profit=df.groupby('State')['Sales','Profit'].sum()
sales_versus_profit.plot.bar(figsize=(15,20))
plt.show()
```

```python
discount=df.groupby('State')['Discount'].sum()
discount
```

```python
plt.figure(figsize=(15,20))
sns.set_style('darkgrid')
discount.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Discount')
```
