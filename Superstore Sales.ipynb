```python
ndf = pd.read_csv('supermarket_sales - Sheet1.csv',parse_dates = ['Date'])
ndf
```

```python
from datetime import datetime as dt
ndf["month"]=ndf['Date'].dt.month
ndf.head(3)
```

```python
ndf["Weekday"]=ndf['Date'].dt.weekday
ndf.head(3)
```

```python
week_df=ndf[ndf["Weekday"]<5]
weekend_df=ndf[ndf["Weekday"]>=5]
print("weekdays sale : ",week_df["Total"].sum())
print("weekends sale : ",weekend_df["Total"].sum())
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['City'] = label_encoder.fit_transform(df['City'])
df['Customer type'] = label_encoder.fit_transform(df['Customer type'])
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Product line'] = label_encoder.fit_transform(df['Product line'])
df['Payment'] = label_encoder.fit_transform(df['Payment'])
```

```python
df.head()
```

```python
x = df.drop(columns =['Invoice ID','Branch','Tax 5%','Time','gross margin percentage','Rating','gross income','Unit price'])
x
```

```python
y = df['gross income']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 10, test_size = 0.3 )
```

```python
x_train
```

```python
x_test
```

```python
linear_model = linear_model.LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
linear_model.score(x_train,y_train)
```

```python
y_pred = linear_model.predict(x_test)
y_pred
```
